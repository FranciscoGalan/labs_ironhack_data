```
# Marcadores

#folium.Marker

#mapa
mapaMarker = folium.Map(location=[39.75568022567029, -104.94703084423678], zoom_start=14)

#direccion
dir = geocoder.osm("2300 Steele St, Denver, CO 80205, United States")

#latitud y longitud
latLong = [dir.lat, dir.lng]

#añadiendo el marcador
folium.Marker(latLong, popup="Bien lejos", tooltip="click").add_to(mapaMarker)

mapaMarker
```

```
#multiples marcadores

dfAirports = vds.airports()
dfAirports.head()
```

```
# creamos el mapa
mapa_aeropuertos = folium.Map(location=[31,-89], zoom_start=6)

# creando los marcadores con iterrows

for indice, fila in dfAirports[:10].iterrows():
  folium.Marker(
      location=[fila.loc['latitude'], fila.loc["longitude"]],
      popup=f"{fila.loc['name']} {fila.loc['city']} {fila.loc['state']}",
      tooltip="click"
  ).add_to(mapa_aeropuertos)

#viendo el mapa
mapa_aeropuertos
```

```
# usando apply

mapa5 = folium.Map(location=[30, -90], zoom_start=5)

#haciendo muchos marcadores
dfAirports[:20].apply(lambda row: folium.Marker(
    location=[row['latitude'], row['longitude']],
    popup=row['name']).add_to(mapa5), axis=1)

#visualizamos
mapa5
```

```
#Más marcadores
#folium.Icon

#creamos el mapa
mapa6 = folium.Map(location=[30, -80], zoom_start=4)

#añadimos un marcador diferente
folium.Marker(location=[30, -80],
              popup='popup',
              icon=folium.Icon(color="green", icon="bolt", prefix='fa')
              ).add_to(mapa6)

#visualizamos el mapa
mapa6
```

```
# ejemplo con muchos iconos diferentes

#creando df random con columna de iconos
df = dfAirports.sample(5)
df = df[['city', 'latitude', 'longitude']]
df['icon'] = ['bicycle', 'car', 'bus', 'truck', 'motorcycle']
df.head()
```

```
# creando el mapa
mapa7 = folium.Map(location=[33, -116], zoom_start=4)

#añadiendo los marcadores
for i in df.itertuples():
  #print(i)
  #print(type(i))
  folium.Marker(
      location=[i.latitude, i.longitude],
      popup=i.city,
      icon=folium.Icon(icon=i.icon, prefix='fa')
  ).add_to(mapa7)

#visualizando el mapa
mapa7
```

```
#creando df random con columna de iconos
df = dfAirports.sample(5)
df = df[['city', 'latitude', 'longitude']]
df['nums'] = [1,2,3,4,5]
df.head()
```

```
# creando el mapa
mapa7 = folium.Map(location=[33, -116], zoom_start=4)

#añadiendo los marcadores
for i in df.itertuples():
  #print(i)
  #print(type(i))
  folium.Marker(
      location=[i.latitude, i.longitude],
      popup=i.city,
      icon=plugins.BeautifyIcon(number=i.nums,
                                border_color="blue",
                                border_width=1,
                                text_color="red",
                                inner_icon_style='margin-top:0px;')
  ).add_to(mapa7)

#visualizando el mapa
mapa7
```

```
#marcadores de circulo

m = folium.Map(
    location=[33, -116],
    zoom_start=13
)
#radio en metros
folium.Circle(
    radius=100,
    location=[33, -116],
    popup='The Waterfront',
    color='crimson',
    fill=False,
).add_to(m)

#radio en pixeles
folium.CircleMarker(
    location=[33, -115],
    radius = 25,
    color="blue",
    fill_color="yellow",
).add_to(m)


m
```

```
# rutas
ciudades = ['Los Angeles', "Las Vegas", "Denver", "Chicago", "Manhattan"]
#dataframe
data = dfAirports[dfAirports.city.isin(ciudades)].drop_duplicates(subset=['city']).sort_values('latitude')
#sacando info de las ciudades
latlongs = data[['latitude', 'longitude']].values
latlongs
```

```
data.head()
```

```
#creando mapa
mapaRutas = folium.Map(location=[32, -92], zoom_start=4)
# añadiendo rutas al mapa
folium.PolyLine(latlongs).add_to(mapaRutas)

#visualizando las rutas
mapaRutas
```

```
#creando mapa
mapaRutas = folium.Map(location=[32, -92], zoom_start=4)
# añadiendo rutas al mapa
plugins.AntPath(latlongs).add_to(mapaRutas)

#visualizando las rutas
mapaRutas
```

```
#Mapa de calor

#vamos a usar geocoder para obtener la información de ciudades

#variables con las coordenadas
mexico = geocoder.osm("Mexico City, Mexico")
monterrey = geocoder.osm("Monterrey, Nuevo Leon")
guadalajara = geocoder.osm("Guadalajara, Guadalajara")

#variables con latlong y una intensidad
cdmx_int = [mexico.lat, mexico.lng, 5000]
monterrey_int = [monterrey.lat, monterrey.lng, 200]
guadalajara_int = [guadalajara.lat, guadalajara.lng, 120]

#lista de las ciudades
ciudades = [cdmx_int, monterrey_int, guadalajara_int]

#creamos el mapa
mapaCalor = folium.Map(location=[19,-99], zoom_start=5)

#añadimos las coordenadas para el mapa de calor
plugins.HeatMap(ciudades).add_to(mapaCalor)

#visualizamos
mapaCalor
```

```
#Mapa de calor con tiempo
plugins.HeatMapWithTime?
```

```
#Generamos coordenadas aleatorias
datos_tiempo = (np.random.random((100,2,2)) + np.array([[19,-99]])).tolist()

#fechas
datos_fechas = [d.strftime("%Y-%m-%d") for d in pd.date_range('20160101', periods=len(datos_tiempo))]

#hacemos el mapa
mapaCalorTiempo = folium.Map([19, -99], zoom_start=6)

#agregamos el layer del mapa de calor
heatmap = plugins.HeatMapWithTime(datos_tiempo, index=datos_fechas)
heatmap.add_to(mapaCalorTiempo)

#visualizamos
mapaCalorTiempo
```
