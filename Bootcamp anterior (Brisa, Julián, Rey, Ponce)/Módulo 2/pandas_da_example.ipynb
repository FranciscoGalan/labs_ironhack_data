```
#lo guardaremos en una variable para eliminar columnas que en su mayoría sean nulos
#el criterio es de cada quien, ver que tan útil nos es la columna y en caso de que no
# eliminarla o ver como recuperar/obtener esta info
func_null = df_final.isnull().sum()
```

```
#eliminamos las columnas y lo guardamos en un nuevo dataframe
df_no_nulls = df_final.drop(cols_to_remove, axis=1)
df_no_nulls.head(1)
```

```
#vamos a ver cuantos nombres de actividad tenemos y como se comportan, queremos ver nuestro top 5 de actividades
df_no_nulls['nombre_act'].unique().tolist()
```

```
#como hay valores en Minusculas y mayusculas, lo cambiamos todo a minusculas y para no tocar la data original lo asignamos a una columna nueva
df_no_nulls['nombre_act_limpio']=df_no_nulls['nombre_act'].str.lower()
```

```
# validando con uno que tiene problemas
df_no_nulls['nombre_act_limpio'][df_no_nulls['nombre_act_limpio'].str.contains("residencias del sector")].unique().tolist()
```

```
#ahora quitaremos signos de puntuación 
puntuacion = [".", "," , "-", "\n", ":", ";",'"', "'", "  ",
                "!", "¡", "¿", "?", "(", ")", "<", "«", "»", "~",]
for p in puntuacion:
                  df_no_nulls['nombre_act_limpio'] = df_no_nulls['nombre_act_limpio'].str.replace(p,"")
```

```
#validando signos de puntuación
df_no_nulls['nombre_act_limpio'][df_no_nulls['nombre_act_limpio'].str.contains("residencias del sector")].unique().tolist()
```

```
#quitando acentos
df_no_nulls['nombre_act_limpio'] =df_no_nulls['nombre_act_limpio'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
```

```
#validando acentos
df_no_nulls['nombre_act_limpio'][df_no_nulls['nombre_act_limpio'].str.contains("residencias del sector")].unique().tolist()
```

```
#sumarizando, creammos un tipo catalogo para hacer un join
df_summarized = df_no_nulls.groupby(['nombre_act_limpio',
                                         'cve_ent']).agg({'id':'count'}).reset_index()
```

```
#validando
df_summarized.head(10)
```

```
#summarized df segun aparicion
df_count_actividad = df_summarized.groupby(['nombre_act_limpio'])['id'].sum()
```

```
#code to remove rows with less than 100 aparitions in the database
rows_to_remove = list(df_count_actividad[df_count_actividad<100].index)
df_count_actividad_clean = df_count_actividad.drop(rows_to_remove,axis=0)
df_count_actividad_clean
```

```
#diferencia entre las actividades... Muchas actividades aparecen muy poco y nos pueden causar ruido
print(df_count_actividad.count())
print(df_count_actividad_clean.count())
```

```
#Top 10 actividades
#agrego to frame para agregar literal de top 10 a los datos
df_summarized_top_10= df_count_actividad_clean.nlargest(10).to_frame()
df_summarized_top_10['class'] = 'top_10'
```

```
#low 10 actividades
#agrego to frame para agregar literal de low 10 a los datos
df_summarized_low_10= df_count_actividad_clean.nsmallest(10).to_frame()
df_summarized_low_10['class'] = 'low_10'
```

```
#contenando y creando catalogo
catalog_top_and_low = pd.concat([df_summarized_top_10,df_summarized_low_10],axis=0).reset_index()
```

```
catalog_top_and_low
```

```
#hacemos el join del catalogo con el dataframe principal
table_concat = pd.merge(df_no_nulls, catalog_top_and_low,how='left', on=['nombre_act_limpio']).reset_index()
```
