Realy Simple Syndication

 Se utiliza para difundir información actualizada frecuentemente a usuarios que se han suscrito a la fuente de contenidos. El formato permite distribuir contenidos sin necesidad de un navegador, utilizando programas llamados agregadores de noticias, diseñados para leer contenidos RSS, tales como Mozilla Firefox, Thunderbird o Akregator, entre otros. A pesar de eso, es posible utilizar el mismo navegador para ver los contenidos RSS.

```python
soup = BeautifulSoup(summary, 'xml')
```

```python
soup
```

```python
rss_feeds_categories = soup.select('div.btnRSS a')

rss_feeds_categories
```

```python
rss_feeds_categories = [feed['href'] for feed in rss_feeds_categories]
```

```python
rss_feeds_categories
```

```python
portada = feedparser.parse(rss_feeds_categories[0])

portada
```

```python
df = pd.DataFrame(portada['entries'])

df.head()
```

```python
df = pd.DataFrame(portada['entries'])

df.head()
```

```python
df_parent = []

for category in rss_feeds_categories:
    try:
        feed = feedparser.parse(category)
        df_child = pd.DataFrame(feed['entries'])
        df_parent.append(df_child)
    except: 
        print(f'We got a problem with this: {category})
        print('The error is:', sys.exc_info()[0])
    
    
```

```python
rss_feeds_categories_clean = []

for category in rss_feeds_categories:
    if " " in category:
        correct_url = category.replace(" ", "%")
        rss_feeds_categories_clean.append(correct_url)
    else:
        rss_feeds_categories_clean.append(category)
        

rss_feeds_categories_clean
        
```

```python
df_parent = []

for category in rss_feeds_categories_clean:
    try:
        feed = feedparser.parse(category)
        df_child = pd.DataFrame(feed['entries'])
        df_parent.append(df_child)
    except: 
        print(f'We got a problem with this: {category})
        print('The error is:', sys.exc_info()[0])
```

```python
len(df_parent)
```

```python
df = pd.concat(df_parent)
```

```python
df.info()
```
