```python
def respuesta(string):
    lista = []

    fake = lambda x: '0' if x < '5' else '1'

    for numero in string:
        lista.append(fake(numero))

    # lista = [fake(numero) for numero in string]

    return ''.join(lista)
```

```python
respuesta('34189471238')
```
