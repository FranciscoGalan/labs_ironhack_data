# Supervised Learning

Aprendizaje supervisado es la parte de Machine Learning que determina una función de datos etiquetados. Un claro ejemplo de desto podría ser el detectar si un correo es SPAM o no, en este caso en el conjunto de datos de entrenamiento con el cual entrnearemos el modelo deberemos pasar los datos con una etiqueta que nos diga cuales son spam y cuales no, esto para que el algoritmo pueda aprender los patrones y  posteriormente identifique en los nuevos inputs patrones simulares para detectar lo más certero posible si es SPAM o no un correo.

Todo este entrenamiento del modelo continuará hasta que se llegue a una precisón buscada y que se considere correcta. Cabe destacar que si en las etiquetas esta mal , el algoritmo no sabrá que esta mal esto y puede que empiece a clasdificar posteriomente esto de manera incorrecta.


Los Tipos de algoritmos más usados en SL son:
>SVM \\
>Linear Regression \\ 
>Logistic Regression \\
>KNearest \\
>Neural Networks

# Unsupervised Learning

A diferencia del aprendizaje supervisado, en el aprendizaje no supervisado se usan datos que no están etiquetados, normalmente pasa esto cuando no conocemos la respuesta al problema que estamos formulando. Esto puede llegar a ser muy incierto ya que no tenemos alguna idea de si nuestro algoritmo esta cerca de la solución adecuada o no. 

Básicamente lo que se intenta hacer con un modelo no supervisado es el que el algoritmo sea capaz de deducir estrucutras y patrones de los datos dados.

Algunos de los métodos más conocidos son:


1.   Clustering: K-Means Clustering
2.   Detección de anomalías
3.   Redes neuronales artificiales



# Reinforcemente Learning

Primero acalararemos conceptos que susaremos:

* Agente = Programa que entrena para hacer el trabajo especificado
* Ambiente = el mundo (real o virtual) en el que el agente realiza acciones
* Acción = Moviemiento del agente
* Recompensas=  Evaluación de una acción, puede ser positiva o negativa

Dados los conceptos, el aprendizaje por refuerzo enseña a un agente a elegir un acción en su propio ambiente, esto con el objetivo de aumentar las recompensas a medida que pasa el tiempo.
