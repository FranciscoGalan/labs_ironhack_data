```python
import matplotlib.pyplot as plt

x_values = range(1, 5001)
y_values = [x**3 for x in x_values]

plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, c=y_values, cmap=plt.cm.Blues, s=2)

#Set chart title and labels
ax.set_title("NÃºmeros cÃºbicos del 1 al 5000", fontsize=22)
ax.set_xlabel('nÃºmero', fontsize=14)
ax.set_ylabel('nÃºmero al cubo', fontsize=14)

#Set the range for each axis.
ax.axis([0, 5000, 0, 125_000_000_000])

# Set size of tick labels.
ax.tick_params(axis='both', labelsize=11)

plt.show()
```

### STacked bar chart

```python
import matplotlib.pyplot as plt

labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 35, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
width = 0.35       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

ax.bar(labels, men_means, width, label='Men')
ax.bar(labels, women_means, width, bottom=men_means,
       label='Women')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.legend()

plt.show()
```
