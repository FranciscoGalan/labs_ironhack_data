## .at y .iat

Son para acceder a valores individuales y modificalos. Son análogos a .loc y .iloc

```python
# El at y iat es mejor para acceder a un elemento y modificarlo. 
george_dupe = pd.Series([10, 7, 1, 22], index=['1968', '1969', '1970', '1970'], name='George Songs')
```

```python
george_dupe
```

```python
george_dupe.at['1970']
```

Para agregar una colmna que use groupby pero sin reducir el tamaño del DataFrame

```python
df_actions_players['media efectividad jugador'] = df_actions_players.groupby('Name').shot_eff.transform("mean")
```
