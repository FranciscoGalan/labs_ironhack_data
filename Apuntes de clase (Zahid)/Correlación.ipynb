# Correlaciones

- Pearson: para relaciones lineales
- Spearman: para relaciones monotónicas (es decir, que si sube una, sube otra, y viceversa). Ver [este enlace](https://www.analyticsvidhya.com/blog/2021/03/comparison-of-pearson-and-spearman-correlation-coefficients/)

```python
import scipy.stats as stats
```

Hacemos pruebas de hipótesis para validar que sí, en efecto, hay una relación y tener sustento de nuestra opinión.

```python
stats.pearsonr(df_actions.Goals, df_actions.ShotsOnfromInsideBox)
```

El 0.0 es el pvalue y te dice qué tan probable es que la hipótesis nula (que no haya relación) es cierta, es decir, que esa correlación se deba a la chiripa. En este caso, no es por chiripa porque no es probable que la hipótesis nula sea cierta. 

```python
serie_corr_kendall_goals = df_actions.corr('kendall')['Goals']
columnas = serie_corr_kendall_goals[(serie_corr_kendall_goals>0.5) | (serie_corr_kendall_goals<-0.5)].index
matriz = df_actions.corr().loc[columnas, columnas]
```

```python
mask = np.zeros_like(matriz, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
plt.subplots(figsize = (15, 10))
cmap = sns.diverging_palette(6, 255, as_cmap=True)

sns.heatmap(matriz, mask=mask, cmap=cmap, center=0, linewidths=.5, annot=True, fmt='.2f');
```
