```python
import plotly.express as px
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = load_breast_cancer()
df_cancer = pd.DataFrame(data = data.data, columns = data.feature_names)
```

```python
df_paises = px.data.gapminder()
df_paises.head()
```

```python
fig = px.scatter(df_paises[df_paises.year == 2007], x = "gdpPercap", y = "lifeExp",
                 size = 'pop', color='continent', log_x = True, hover_name = 'country', size_max = 60)
fig.show()
```

```python
fig = px.scatter(df_cancer, x = 'mean radius', y='mean perimeter',
                 size='mean area', color = 'mean texture')
fig.show()
```

```python
plt.figure(figsize=(12,6))
fig = px.scatter_3d(df_cancer, x = 'mean radius', y ='mean perimeter', z = 'mean area',
                    color = 'mean texture', opacity = 0.5)
fig.show()
```
